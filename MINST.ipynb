```python
train_path = ("C:\\Users\\npram\\.kaggle\\Data_download\\competitions\\Digit-Recognizer\\Digit-Recognizer\\train.csv")
test_path = ("C:\\Users\\npram\\.kaggle\\Data_download\\competitions\\Digit-Recognizer\\Digit-Recognizer\\test.csv")
```

```python
import pandas as pd
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
print(train.shape)
train.head()
```

```python
print(test.shape)
test.head()
```

```python
# Input image data

# 1. Covert the data from dataframe to nupy array to effiticely store the image data

import numpy as np

X = np.array(train.drop("label", axis=1))
X = X.reshape(-1,28,28,1)    # reshaping into grey scale
X = X/255.0     # Regularization by dividing the pixels so as to restrict them between 0 and 1
print(X)
Y = pd.get_dummies(train.label)
Y = np.array(Y)

print("Oraginal data shape: ", train.shape)
print("Reshaping the X data: ", X.shape)
print("Shape of the Y data: ", Y.shape)
```

```python
import matplotlib.pyplot as plt

fig,plot = plt.subplots(2, 5, figsize=(16,7))

for i in range(10):
    z = np.random.randint(low=0, high=X.shape[0])
    axes = plot[i//5, i%5]
    print(axes)
    axes.axis("off")
    axes.imshow(X[z].reshape(28,28), cmap='gray')
    axes.set_title("Label : " + str(np.argmax(Y[z])))
    print("Label : " + str(np.argmax(Y[z])))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator 

train_x, val_x, train_y, val_y = train_test_split(X, Y, test_size=0.2, random_state=40)

generator = ImageDataGenerator(rotation_range=45, 
                               width_shift_range=0.2, 
                               height_shift_range=0.2,
                               zoom_range=0.2, 
                               shear_range=0.2)
train_values = generator.flow(train_x, train_y)
val_values = generator.flow(val_x, val_y)
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D, BatchNormalization

model = Sequential([Conv2D(32, kernel_size=(3,3), activation=tf.nn.relu, padding='Same', input_shape=(28,28, 1)),
                    MaxPooling2D(pool_size=(2,2)),
                    Dropout(0.2),
                    Conv2D(32, kernel_size=(3,3), padding='Same', activation=tf.nn.relu),
                    MaxPooling2D(pool_size=(2,2)),
                    BatchNormalization(),
                    Flatten(),
                    Dense(64, activation=tf.nn.relu),
                    BatchNormalization(),
                    Dropout(0.2),
                    Dense(32, activation=tf.nn.relu),
                    Dropout(0.2),
                    Dense(10, activation=tf.nn.softmax)])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), 
              loss=tf.keras.losses.CategoricalCrossentropy(), 
              metrics=['accuracy'])
callback = EarlyStopping(monitor="val_accuracy", mode="max", patience=15, verbose=1)
history = model.fit(train_values, validation_data=val_values, batch_size=32, 
                    epochs=100, verbose=1, callbacks = [callback])
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs= range(len(acc))

plt.figure(figsize=(16, 5))
plt.subplot(121)
plt.plot(epochs, acc, label='Train_acc')
plt.plot(epochs, val_acc, label='Val_acc')
plt.legend()
plt.title('Accuracy')

plt.subplot(122)
plt.plot(epochs, loss, label='Train_loss')
plt.plot(epochs, val_loss, label='Val_loss')
plt.legend()
plt.title('Loss')
plt.show()
```

```python
test_values = np.array(test)/255.0
test_values = test_values.reshape(-1, 28, 28, 1)
print("Test data shape : " + str(test_values.shape))
pred = np.argmax(model.predict(test_values), axis=1)

sub = pd.DataFrame({'ImageId': list(range(1, pred.shape[0]+1)), 'Label':pred})
sub.to_csv('./submission.csv', index=False)
```
